```python
#prime number
num=6
if num>1:
    for i in range(2,num):
        if num%i==0:
            print(num,'is not a prime number')
            break
    else:
        print(num,'is a prime number')
else:
    print(num,'is not a prime number')
```

```python
#factorial number
num=int(input("Enter the number:"))
fact=1
if num<0:
    print("Factorial does not exist for the negative numbers")
elif num==0:
    print('factorial of 0 is 1')
else:
    for i in range(1,num+1):
        fact=fact*i
        print("The factorial of",num,'is',fact)
    
```

```python
import math
num=int(input("Enter the number: "))
if num<0:
    print("factorial does't exist negative numbers")
else:
    print("The factorial of",num,'is',math.factorial(num))
```

```python
#factorial using while loop
num=int(input("Enter the number:"))
fact=1
i=1
if num<0:
    print("Factorial does not exist for the negative numbers")
elif num==0:
    print('factorial of 0 is 1')
else:
    while i<=num:
        fact*=i
        i+=1
        print("The factorial of",num,'is',fact)
    
```

```python
#Amstrong number
num=int(input("Enter the number:"))
n=len(str(num))
sum_of_powers=sum(int(digit)**n for digit in str(num))
if num==sum_of_powers:
    print(num,'is armstrong number')
else:
    print(num,'is not a armstrong number')
```
